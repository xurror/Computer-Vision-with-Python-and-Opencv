```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../Computer-Vision-with-Python/DATA/dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
plt.imshow(cv2.flip(fix_img,-1))
```

```python
plt.imshow(cv2.rectangle(fix_img,(195,370),(600,720),(255,0,0),5))
```

```python
vertices = np.array([[270,720], [500,370], [730,720]], dtype=np.int32)
pts = vertices.reshape((-1,1,2))
plt.imshow(cv2.polylines(fix_img,[pts],isClosed=True,color=(0,0,255),thickness=20))
```

```python
plt.imshow(cv2.fillPoly(fix_img,[pts],(0,0,255)))
```
